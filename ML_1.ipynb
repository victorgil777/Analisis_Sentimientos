```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def CargaDatos():
    df = pd.read_excel('DatosRegresion1D.xlsx')
    x = df['ingresos']
    y = df['consumo']
    return x,y
```

```python
def DibujaDatos(x,y,n):
    plt.figure()
    plt.plot(x[:n],y[:n],'x')
    plt.xlabel('Ingresos anuales (miles de â‚¬)')
    plt.ylabel('Consumo anual de electricidad (Mw-h)')
```

```python
x,y = CargaDatos() #Apartado b
DibujaDatos(x,y,100)  #Apartado c
```
